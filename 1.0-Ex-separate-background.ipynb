```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib qt   # mask image pop up
```

```python
path='D:/AI epclon/udacity CV/Udacity - Computer Vision Nanodegree/Udacity - Computer Vision Nanodegree nd891 v1.0.0/Part 01-Module 01-Lesson 02_Image Representation & Classification/img'
```

```python
image=cv2.imread(path+'/fisih.jpg')
plt.imshow(image);
```

```python
image_copy=np.copy(image)
image_copy=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy);
```

```python
image.shape
```

```python
Green_min_threshold=np.array([0,240,0])
Green_max_threshold=np.array([100,255,100])
mask=cv2.inRange(image_copy,Green_min_threshold,Green_max_threshold)
plt.imshow(mask);
```

```python
plt.imshow(mask, cmap='gray');
```

```python
masked_image=np.copy(image_copy)
masked_image[ mask !=0 ]=[0,0,0]
plt.imshow(masked_image);
```

```python
background_image=cv2.imread(path+'/sea.jpg')
background_image=cv2.cvtColor(background_image,cv2.COLOR_BGR2RGB)
background_image=cv2.resize(background_image,(1280,720))
plt.imshow(background_image);
```

```python
back_copy=np.copy(background_image)
back_copy[mask == 0] =[0,0,0]
plt.imshow(back_copy);
```
