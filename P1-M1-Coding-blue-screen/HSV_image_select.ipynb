```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#load pallons image 
path='D:/AI epclon/udacity CV/Udacity - Computer Vision Nanodegree/Udacity - Computer Vision Nanodegree nd891 v1.0.0/Part 01-Module 01-Lesson 02_Image Representation & Classification/img'
#image=cv2.imread(path+'/pallons1.jpg')
#image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image=plt.imread(path+'/pallons1.jpg')
plt.imshow(image);
```

```python
r=image[: ,: ,0]
g=image[: ,: ,1]
b=image[: ,: ,2]
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.imshow(r , cmap='gray')
plt.title('red image ')
plt.subplot(1,3,2)
plt.imshow(g , cmap='gray')
plt.title('green  image ')
plt.subplot(1,3,3)
plt.imshow(b , cmap='gray')
plt.title('blue image ')
```

```python
image_copy=np.copy(image)
HSV=cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)
plt.imshow(HSV);
```

```python
h=HSV[: ,: ,0]
s=HSV[: ,: ,1]
v=HSV[: ,: ,2]
```

```python
f,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,20))

ax1.imshow(h , cmap='gray')
ax1.set_title('hue image ')

ax2.imshow(s, cmap='gray')
ax2.set_title('saturation  image ')

ax3.imshow(v , cmap='gray')
ax3.set_title('value image ');
```

```python
pink_lower=np.array([180, 0, 100])
pink_upper=np.array([255, 255, 230])

hue_lower=np.array([120,0,0])
hue_upper=np.array([180,250,250])
```

```python
mask=cv2.inRange(image,pink_lower,pink_upper)
masked_image=np.copy(image)
masked_image[mask==0]=[0,0,0]
plt.imshow(masked_image )
```

```python
mask=cv2.inRange(HSV,hue_lower,hue_upper)
masked_image=np.copy(image)
masked_image[mask==0]=[0,0,0]
plt.imshow(masked_image )
```
